# Zadanie 1: Klasteryzacja klientów na podstawie ich zakupów


Użyj zbioru danych "Mall Customers" z Kaggle, który zawiera informacje o klientach centrum handlowego (ID klienta, wiek, płeć, dochód roczny, ocena wydatków).**Zastosuj algorytm K-Means, aby podzielić klientów na segmenty na podstawie ich dochodu rocznego i oceny wydatków.**


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Załadowanie zbioru danych
file_path  = 'Mall_Customers.csv'
data = pd.read_csv(file_path)

# Wybór cech do klasteryzacji
X = data[['Annual Income (k$)', 'Spending Score (1-100)']]

# Zastosowanie K-Means
kmeans = KMeans(n_clusters=2, n_init=10)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Wizualizacja klastrów
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title("K-Means Clustering on Mall Customers")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.show()
```
